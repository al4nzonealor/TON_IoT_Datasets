```python
# Data handling and EDA tools
import pandas as pd
import numpy as np

import seaborn as sns
import plotly.express as px
from matplotlib import pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

#Display Aid
from IPython.display import display
```

```python
# Read all the 7 csv files into 7 different dataframes
fridge = pd.read_csv('./Train_Test_datasets/Train_Test_IoT_Fridge.csv')
garage = pd.read_csv('./Train_Test_datasets/Train_Test_IoT_Garage_Door.csv')
gps_tracker = pd.read_csv('./Train_Test_datasets/Train_Test_IoT_GPS_Tracker.csv')
modbus = pd.read_csv('././Train_Test_datasets/Train_Test_IoT_Modbus.csv')
motion_light = pd.read_csv('./Train_Test_datasets/Train_Test_IoT_Motion_Light.csv')
thermo = pd.read_csv('./Train_Test_datasets/Train_Test_IoT_Thermostat.csv')
weather = pd.read_csv('./Train_Test_datasets/Train_Test_IoT_Weather.csv')
```

```python
# Take a Look At the Shape and the Head of each Dataframe
dataframe_lst = [fridge, garage, gps_tracker, modbus, motion_light, thermo, weather]
title_lst = ["Fridge Dataset", "Garage Door Dataset", "GPS Tracker Dataset", "Modbus Dataset", "Motion Light Dataset", "Thermostat Dataset", "Weather Dataset"]
for dataframe_name, title in zip(dataframe_lst, title_lst):
    print("================")
    print(f"Head of {title}")
    print("================")
    print(f"Shape = {dataframe_name.shape}")
    display(dataframe_name.head())
```

As it can be seen from the execution of the above cell all the datasets consisted of columns date, ts and time which are related to the time of recording the data attributes specific to the dataset. Also, we can see that each of the datasets contain two columns namely "label" which describes the type of the system behaviour 1 for anomalous(Attck behavior) and 0 for Normal system behavior and "type" describes which specific type of cyber attack is there in specific device for the specific observed system behaviors.

```python
for dataframe_name, title in zip(dataframe_lst, title_lst):
    print(f"Info of {title}")
    display(dataframe_name.info())
```

As we can see the execution of the cell above All of our datasets contain non-null values in which we are not expected to be treating missing values(impute/remove).

```python
# Upon various tests run on each of the datasets I have found the latitude and longitude columns of the gps_tracker dataframe to
# have errounous values which are extreamly high thus I have tried to truncate them to the normal range
# Latitude is specified in degrees within the range [-90, 90]. Longitude is specified in degrees within the range [-180, 180]
gps_tracker = gps_tracker[gps_tracker['longitude'] <= 180]
```

```python
for dataframe_name, title in zip(dataframe_lst, title_lst):
    print(f"Cyber Attack for Different 'Date' in {title}")
    display(dataframe_name.groupby('date')['type'].unique().reset_index())
```

The execution of the cell above shows the distribution of different cyber attacks over different dates in our datasets for each IOT device type.

```python
# Unique Feature List in each dataset apart from ['ts', 'date', 'time', 'label', 'type']
time_related = ['ts', 'date', 'time', 'label', 'type']

count = 0
col_name_list = []
for dataframe_name, title in zip(dataframe_lst, title_lst):
    col_list = dataframe_name.columns
    print("=================")
    print(f"****{title}****")
    print("=================")
    for j in col_list:
        if j not in time_related:
            print(j)
            col_name_list.append(j)
            count += 1

print(f"Count of all features across all datasets: {count}")
```

```python
for x in ['label','type']:
    col_name_list.append(x)
```

```python
# Creating a bigger dataframe that can merge all our data for all IOT component types
df = pd.DataFrame(columns=col_name_list)
imp_columns_per_iot_component = {'fridge': ['fridge_temperature', 'temp_condition','label','type'], 
               'garage': ['door_state', 'sphone_signal','label', 'type'], 
               'gps': ['latitude', 'longitude','label', 'type'],
               'modbus': ['FC1_Read_Input_Register','FC2_Read_Discrete_Value', 'FC3_Read_Holding_Register', 'FC4_Read_Coil','label', 'type'], 
               'motion_light': ['motion_status', 'light_status','label', 'type'],
               'thermo': ['current_temperature', 'thermostat_status','label', 'type'],
               'weather': ['temperature', 'pressure', 'humidity','label','type']
               }

iot_componenet_level_df = []

for dataset_name, component_name in zip(dataframe_lst, list(imp_columns_per_iot_component.keys())):
    temp_component_level_df = dataset_name.drop(columns='ts')

    temp_df = pd.DataFrame(columns=col_name_list)

    temp_df[imp_columns_per_iot_component[component_name]] = temp_component_level_df.loc[:, imp_columns_per_iot_component[component_name]]

    temp_df = temp_df.fillna(-1).reset_index(drop=True)
    print(temp_df.head())
    
    df = pd.concat([df, temp_df])
```

```python
# Save the final dataframe(for training the models)
df.to_csv('./final.csv', index=False)
```

```python
df = pd.DataFrame(columns=col_name_list)
imp_columns_per_iot_component = {'fridge': ['fridge_temperature', 'temp_condition','iot_component_type', 'label','type'], 
               'garage': ['door_state', 'sphone_signal','iot_component_type', 'label', 'type'], 
               'gps': ['latitude', 'longitude','iot_component_type', 'label', 'type'],
               'modbus': ['FC1_Read_Input_Register','FC2_Read_Discrete_Value', 'FC3_Read_Holding_Register', 'FC4_Read_Coil','iot_component_type','label', 'type'], 
               'motion_light': ['motion_status', 'light_status','iot_component_type','label', 'type'],
               'thermo': ['current_temperature', 'thermostat_status','iot_component_type','label', 'type'],
               'weather': ['temperature', 'pressure', 'humidity','iot_component_type','label','type']
               }

iot_componenet_level_df = []

for dataset_name, component_name in zip(dataframe_lst, list(imp_columns_per_iot_component.keys())):
    temp_component_level_df = pd.DataFrame(dataset_name.groupby('ts').apply(lambda x: x.mode().iloc[0])).drop(columns='ts')
    temp_component_level_df['iot_component_type'] = component_name

    temp_df = pd.DataFrame(columns=col_name_list)

    temp_df[imp_columns_per_iot_component[component_name]] = temp_component_level_df.loc[:, imp_columns_per_iot_component[component_name]]

    temp_df = temp_df.fillna(0).reset_index(drop=True)
    print(temp_df.head())
    
    df = pd.concat([df, temp_df])
```

```python
# Save the final dataframe(for further exploratory data analysis since this dataframe have a columns called iot_component_type 
# I actually will be using it alot)
df.to_csv('./final_df.csv', index=False)
```
