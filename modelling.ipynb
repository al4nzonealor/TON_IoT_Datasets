#### Discussion 2
Before training any model let's first handle categorical variables in our dataset and convert them to equivalent numerical representaion(encoding). For columns containg only two categories [column names: 'light_status', 'sphone_signal', 'door_state', 'temp_condition'] I will be using binary encoding.

One-hot Encoding: One-hot encoding is a categorical variable encoding technique that transforms categorical data into a binary representation while maintaining the distinctiveness of categories. It involves creating separate binary columns, for each unique category within the column. Each category is assigned a separate column, and the presence or absence of a specific category is denoted by a binary value of 1 or 0, respectively. One-hot encoding ensures that the encoded features are mutually exclusive, without implying any ordinal relationships between the categories.

```python
df.columns
```

```python
col_to_encode = ['motion_status', 'light_status', 'sphone_signal', 'door_state', 'temp_condition']
final_df = df

for i in col_to_encode:
    final_df = pd.concat([final_df, pd.get_dummies(final_df[i], prefix=i)], axis=1)
    final_df = final_df.drop(columns=i)
```

#### Discussion 3
Since we have now converted all our non-numeric values into numeric we can start training the bench mark model

A benchmark model, refers to a baseline model that serves as a reference point for evaluating the performance of other, more complex models. It provides a standard against which the performance of new models can be compared. A benchmark model is typically simple and easy to implement, allowing for quick evaluation and comparison of more advanced techniques. It helps establish a minimum performance level that should be exceeded by any new model to be considered valuable.

I have choosen logistic regression as a bench mark model for classifying cyber types of cyber attacks on verious IoT component devices such as Fridge, Garage Door, Modbus. I picked logistic regression since it is widely used and well-understood algorithm for binary classification. It can be extended to handle multiclass problems through techniques like one-vs-rest or multinomial logistic regression. Also, Logistic regression is computationally efficient and relatively interpretable compared to more complex models like neural networks or ensemble methods. This simplicity enables the benchmark model to act as a reference point for assessing the performance of more sophisticated algorithms.

#### Bench Mark Model - Logistic Regression

```python
# Split the dataframe into input features X and target_features y
X = final_df[
    ['fridge_temperature', 'latitude', 'longitude',
       'FC1_Read_Input_Register', 'FC2_Read_Discrete_Value',
       'FC3_Read_Holding_Register', 'FC4_Read_Coil', 'current_temperature',
       'thermostat_status', 'temperature', 'pressure', 'humidity', 
       'motion_status_-1', 'motion_status_one', 'motion_status_zero',
       'light_status_-1', 'light_status_off', 'light_status_on',
       'sphone_signal_-1', 'sphone_signal_false', 'sphone_signal_true',
       'door_state_-1', 'door_state_closed', 'door_state_open',
       'temp_condition_-1', 'temp_condition_high', 'temp_condition_low']
]
y = final_df['type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state = 1)

print('X_train dimension= ', X_train.shape)
print('X_test dimension= ', X_test.shape)
print('y_train dimension= ', y_train.shape)
print('y_train dimension= ', y_test.shape)
```

Since we are going to use One Vs Rest algorithm, set > multi_class='ovr'
Note: since we are using One Vs Rest algorithm we must use 'liblinear' solver with it

```python
# Train our model over the training dataset
lm = LogisticRegression(multi_class='ovr', solver='liblinear')
lm.fit(X_train, y_train)
```

```python
# Record the performance of the model
lm.score(X_test, y_test)
```

#### Confusion Matrix For The Bench Mark Model
Confusion matrix helps to visualize the performance of the model. The diagonal elements of the represent the number of points for which the predicted label is equal to the true label Off-diagonal elements are those that are mislabeled by the classifier. The higher the diagonal values of the confusion matrix the better, indicating many correct classifications by the classifier.

```python
print(metrics.classification_report(y_test, lm.predict(X_test)))
```

```python
class_labels = ['ddos', 'backdoor', 'injection', 'normal', 'password','ransomware', 'xss', 'scanning']

fig, ax = plt.subplots(figsize=(10, 8))

confusion_matrix_log = np.log10(metrics.confusion_matrix(y_test, lm.predict(X_test), labels=class_labels) + 1)

sns.heatmap(confusion_matrix_log, annot=True, fmt='.1f', cmap='Blues', cbar=False, xticklabels=class_labels, yticklabels=class_labels, ax=ax)

ax.set_xlabel('Predicted Class')
ax.set_ylabel('True Class')
ax.set_title('Confusion Matrix')

plt.show()
```

#### Decision Tree Classifier

Decision tree is a type of supervised learning algorithm (having a predefined target variable) that is mostly used in classification problems. It works for both categorical and continuous input and output variables. In this technique, we split the population or sample into two or more homogeneous sets (or sub-populations) based on most significant splitter / differentiator in input variables. A decision tree is a flowchart-like tree structure where an internal node represents feature(or attribute), the branch represents a decision rule, and each leaf node represents the outcome. The topmost node in a decision tree is known as the root node. It learns to partition on the basis of the attribute value. It partitions the tree in recursively manner call recursive partitioning. This flowchart-like structure helps you in decision making. It's visualization like a flowchart diagram which easily mimics the human level thinking. That is why decision trees are easy to understand and interpret. Decision Tree is a white box type of ML algorithm. It shares internal decision-making logic, which is not available in the black box type of algorithms such as Neural Network. Its training time is faster compared to the neural network algorithm. The time complexity of decision trees is a function of the number of records and number of attributes in the given data. The decision tree is a distribution-free or non-parametric method, which does not depend upon probability distribution assumptions. Decision trees can handle high dimensional data with good accuracy.

```python
from sklearn.tree import DecisionTreeClassifier

model_dt1 = DecisionTreeClassifier(random_state=42)
model_dt1.fit(X_train, y_train)

print('Depth of the Decision Tree: ', model_dt1.get_depth())
print('Accuracy on training: ',model_dt1.score(X_train, y_train))

yhat1 = model_dt1.predict(X_train)
print(yhat1)

from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(y_train.values, yhat1)
print('-------The confusion matrix for this model is-------')
print(cm1)

from sklearn.metrics import classification_report
print('\n\n-------Printing the whole report of the model-------')
print(classification_report(y_train.values, yhat1))

class_labels = ['ddos', 'backdoor', 'injection', 'normal', 'password','ransomware', 'xss', 'scanning']

fig, ax = plt.subplots(figsize=(10, 8))

sns.heatmap(confusion_matrix(y_train.values, yhat1, labels=class_labels), annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels, ax=ax)

ax.set_xlabel('Predicted Class')
ax.set_ylabel('True Class')
ax.set_title('Confusion Matrix')
```

For the Cell above we have tried to plot the confusion matrix for the performance of our model in the training dataset where the classification
report contains very widely varying ranges of values which makes the color range of the chart very poor so to mitigate this effect we will be using "Logarithimic Scaling" on the values of the confusion matrix.

```python
confusion_matrix_log = np.log10(confusion_matrix(y_train.values, yhat1, labels=class_labels) + 1)
fig, ax = plt.subplots(figsize=(10, 8))

sns.heatmap(confusion_matrix_log, annot=True, fmt='.1f', cmap='Blues', cbar=False, xticklabels=class_labels, yticklabels=class_labels, ax=ax)

ax.set_xlabel('Predicted Class')
ax.set_ylabel('True Class')
ax.set_title('Confusion Matrix')

plt.show()
```

#### Randomforest Classifier

Random forest is a supervised learning algorithm. It has two variations – one is used for classification problems and other is used for regression problems. It is one of the most flexible and easy to use algorithm. It creates decision trees on the given data samples, gets prediction from each tree and selects the best solution by means of voting. It is also a pretty good indicator of feature importance.
Random forest algorithm combines multiple decision-trees, resulting in a forest of trees, hence the name Random Forest. In the random forest classifier, the higher the number of trees in the forest results in higher accuracy.
